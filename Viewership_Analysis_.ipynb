# Create watch time buckets

```python
df.describe()
```

```python
# Define the bin edges for 60-minute intervals (which is 1 hour)
bins = np.arange(0, df['TotalTimeWatched_minutes'].max() + 60, 60)

# Create labels for the bins in hours
labels = [f'< 1 hour'] + [f'{int(i/60)}-{int((i+60)/60)} hours' for i in bins[1:-1]]

# Handle the last bin label separately
if len(labels) < len(bins) - 1:
    labels.append(f'{int(bins[-2]/60)}+ hours')


# Create a new column with the watch time buckets
df['WatchTime_Buckets'] = pd.cut(df['TotalTimeWatched_minutes'], bins=bins, labels=labels, right=False, include_lowest=True)

# Display the value counts for the new column
display(df['WatchTime_Buckets'].value_counts().sort_index())
```
